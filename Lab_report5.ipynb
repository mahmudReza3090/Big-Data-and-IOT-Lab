```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("credit card.csv")
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
fill= df['MARRIAGE'].median()
df.AGE = df.AGE.fillna(fill)
df.AGE
```

```python
X = df.drop(['default.payment.next.month'],axis=1).fillna(0)
X
```

```python
y= df['default.payment.next.month'].fillna(0)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=1)
```

```python
import seaborn
from sklearn.svm import SVC
r=SVC(gamma='auto')
r.fit(X_train, y_train)
```

```python
pred1= r.predict(X_test)
```

```python
r.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import classification_report
accuracy_score(y_test,pred1)
```

```python
confusion_matrix(y_test,pred1)
```

```python
classification_report(y_test,pred1)
```

```python
print(classification_report(y_test,pred1))
```

```python
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = numpy.random.binomial(1,.77,size = 30000)
predicted = numpy.random.binomial(1,.78,size = 30000)

confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show()
```
